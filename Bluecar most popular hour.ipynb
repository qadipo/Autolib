```
# loading the required libraries
import pandas as pd
import numpy as np
```

```
# loading the dataset
autolib = pd.read_csv('Autolib_dataset (2).csv')
```

```
# snip view of the data
autolib.head(10)
```

```
# information about the data
autolib.info()
```

```
# Dropping columns
# Reason: May not be useful in our analysis
autolib = autolib.drop(['Cars', 'Charging Status', 'Scheduled at', 'Displayed comment', 'Status', 'Kind'], axis=1)
```

```
#Combining year, month, day, hour and minutes colums to create datetime
autolib['year'] = autolib['year'].astype(str)
autolib['month'] = autolib['month'].astype(str)
autolib['day'] = autolib['day'].astype(str)
autolib['hour'] = autolib['hour'].astype(str)
autolib['minute'] = autolib['minute'].astype(str)

autolib['DateTime'] = autolib['year']+'-'+autolib['month']+'-'+autolib['day']+' '+autolib['hour']+':'+autolib['minute']

autolib['DateTime'] = pd.to_datetime(autolib['DateTime'])

```

```
# Dropping the year, month, day, hour, minutes columns
autolib = autolib.drop(['year','month','day','minute','Geo point','Charge Slots'], axis=1)
```

```
autolib.info()
```

```
# Checking any missing values
autolib.isnull().sum()
```

```
# Removing duplicate values if any
autolib.drop_duplicates()
```

```
# Filtering the data
# Focus is on Paris city for the month of April 2018
autolib_paris = autolib[(autolib.City == 'Paris') & (autolib.DateTime >= '2018-04-01') & (autolib.DateTime <= '2018-04-30')]
autolib_paris.head(10)
```

```
# Further filtering out stations that are operational only
autolib_paris = autolib_paris[autolib_paris['Rental status'] == 'operational']
autolib_paris.head(10)
```

```
# Sorting the data by station and date for subsequent operations
# The Postal code is unique for each so it has been considered to 
autolib_paris_sorted = autolib_paris.sort_values(by=['Public name','DateTime'], ascending=True)
```

```
# Snip view of the data
autolib_paris_sorted.head(10)
```

```
# Computing usage per station
autolib_paris_sorted['Usage'] = autolib_paris_sorted['Bluecar counter'].diff()
```

```
# Viewing the data
autolib_paris_sorted.head(10)
```

**ANALYSIS**

```
#QUESTION ONE: Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018
autolib_paris_sorted[autolib_paris_sorted.Usage < 0].groupby('hour')['Usage'].sum().sort_values().head(1)
```

```
# What is the most popular hour for returning cars?
# Return ---> values greater than 0
# Picking ---> values less than 0
autolib_paris_sorted[autolib_paris_sorted.Usage > 0].groupby('hour')['Usage'].sum().sort_values().head(1)
```
